```python
import numpy as np
from PIL import Image
import os
rootDir='photos'
arr=np.zeros([20,200,200,3])
def main():
    try:
        def num_of_files():
            try:
                for dirname,subdirList,fileList in os.walk(rootDir):
                    file_Count=len(fileList)
                    print("Total No Of file = ",file_Count)
            except FileNotFoundError:
                print("File not found")
        def find_dir():
            try:
                for dirname,subdirList,fileList in os.walk(rootDir):
                    print("Found directory: %s" %dirname)
                    for fname in fileList:
                        print("\t%s"%fname)
            except FileNotFoundError:
                print("File not found")
        def img_resize():
            try:
                i=0
                for dirname,subdirList,fileList in os.walk(rootDir):
                    for fname in fileList:
                        filePath=dirname+os.sep+fname
                        if filePath.endswith('.jpg'):
                            img=Image.open(filePath)
                            img_n=img.resize((200,200))
                            arr[i]=img_n
                            i+=1
                print(arr)
            except FileNotFoundError:
                print("File not found")
        print(num_of_files())
        print(find_dir())
        print("Numpy array of resized image")
        print(img_resize())
    except RuntimeError:
        print("Found an error...!")
if __name__ == '__main__':
    main()
        
```
